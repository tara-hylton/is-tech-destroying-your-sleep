Creating databases from the 3 csv files

```python
data1 = pd.read_csv('raw_data/anxiety_depression_data.csv')
data2 = pd.read_csv('raw_data/mental_health_and_technology_usage_2024.csv')
data3 = pd.read_csv('raw_data/brain_rot_cases.csv')
data4 = pd.read_csv('raw_data/smmh.csv')
```

Removed unwanted columns

```python
data1 = data1.drop(columns=['Substance_Use', 'Financial_Stress', 'Self_Esteem_Score', 'Life_Satisfaction_Score', 'Loneliness_Score','Family_History_Mental_Illness', 'Medication_Use', 'Therapy', 'Meditation'])
data1
```

Removed unwanted columns

```python
data2 = data2.drop(columns=['User_ID', 'Social_Media_Usage_Hours', 'Gaming_Hours', 'Screen_Time_Hours', 'Online_Support_Usage'])
data2
```

Updated column names, removed unwanted columns. Changed columns that were integers, such as gender and education, to text to make them comparable to other datasets.

```python
data3 = data3.rename(columns={'How many hours do you spend daily on social media': 'Hours_per_day_on_social_media', 'Do you prefer using technology for work or leisure?': 'NA1', 'How many hours do you spend in front of computer screens or smartphones daily?': 'Hours_per_day_using_technology', 'Do you feel that you use social media excessively?': 'NA2', 'Do social media platforms affect your social relationships?': 'NA3', 'Have you noticed a change in your mood after using technology?': 'Mood_change_after_use', 'How would you rate your level of focus while working or studying? On a scale from 1 to 5, where 1 = very poor and 5 = excellent.': 'NA4', 'Do you find it difficult to concentrate when using technology (social media)?': 'NA5', 'How often do you feel distracted while trying to work or study due to technology (social media)?': 'NA6', 'How would you rate your level of concern regarding technology use, particularly social media? On a scale from 1 to 5, where 1 = very low and 5 = excellent.': 'NA7', 'Do you believe that your use of social media negatively affects y How would you rate your level of concern regarding technology use, particularly social media? On a scale from 1 to 5, where 1 = very low and 5 = excellent.our mental health?': 'NA8', 'Do you believe that your use of social media negatively affects your mental health?': 'Does_social_media_negatively_affect_you?', 'What platforms do you primarily use?': 'NA9', 'Do you think that using certain applications can help reduce feelings of mental fatigue?': 'Do_certain_apps_help_mental_fatigue?', 'Do you use any strategies to reduce technology usage?': 'Do_you_try_to_reduce_technology_usage?', 'If yes was chosen for the above question, what are the strategies?': 'Strategies_to_reduce_technology_use', 'Educational Level': 'Education'})
data3 = data3.drop(columns=['NA1', 'NA2', 'NA3', 'NA4', 'NA5', 'NA6', 'NA7', 'NA8', 'NA9'])
data3['Gender'] = data3['Gender'].replace({0: 'Female', 1: 'Male'})
column_list3 = ['Mood_change_after_use', 'Does_social_media_negatively_affect_you?', 'Do_certain_apps_help_mental_fatigue?', 'Do_you_try_to_reduce_technology_usage?']
for column in column_list3:    
    data3[column] = data3[column].replace({0: 'No', 1: 'Yes'})
data3['Education'] = data3['Education'].replace({0:'Did not finish high school', 1: 'HS Diplomoa or equivalent', 2: "Associate's Degree", 3: "Bachelor's Degree", 4: 'Post Bachelor Studies'})

data3
```
